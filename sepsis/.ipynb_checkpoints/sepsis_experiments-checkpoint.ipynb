## Sepsis Experiments

This notebook is used to generates plots for sepsis experiment in section 6. To run this notebook the following data is necessary which can be obtained by either running `learn_policies.ipynb` or `uzip` the file `data/processed.zip`. Make sure the following list of file exists in the `data` directory:

1. `optimal_policy_st.pkl`
2. `mixed_policy.pkl`
3. `t0_policy.pkl`
4. `value_function.pkl`
5. `tx_tr.pkl`

The strcutre of this notebook is as follows,

1. Generate the data: First section used the `conf_data_generator` class in `core` to simulate 10000 trajectories.
2. Weighted Importance Sampling estimate: This section estimates the WIS estimate of each policies `optimal_policy_st`, `with antibiotics` and `without antibiotics`
3. Our method and the naive estimate: Computes the lower/ upper bound using our method and the naive method.

````python
import numpy as np
import pickle
from tqdm import tqdm

from core.sepsisSimDiabetes.State import State
from core.sepsisSimDiabetes.Action import Action
from core import generator_confounded_mdp as DGEN
from core import conf_wis as CWIS
from core import loss_minimization as LB
from utils.utils import *

import seaborn as sns
sns.set_style("whitegrid")

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
figdir='./figs'
````

### Loading the preprocessed data

````python
with open('data/value_function.pkl', 'rb') as f:
    value_function = pickle.load(f)
with open('data/t0_policy.pkl', 'rb') as f:
    t0_policy = pickle.load(f).transpose([0, 2, 1])
with open('data/optimal_policy_st.pkl', 'rb') as f:
    optimal_policy_st = pickle.load(f).transpose([1, 0])
with open('data/tx_tr.pkl', 'rb') as f:
    tx, tr = pickle.load(f)
with open('data/mixed_policy.pkl', 'rb') as f:
    mixed_policy = pickle.load(f).transpose([1, 0])
````

### Data Generation

**config:** the config file containts the following informations:

- `Gamma`: amount of confounding in the data generation process ($\Gamma^\star$)
- `num_itrs`: number of simulation trajectories
- `max_horizon`: maximum number of timesteps in the sim
- `discount`: discount factor in the MDP ($\gamma$)
- `confounding_threshold`: confounding threshold (refer to Appendix D)
- `nS`: number of states (including the termina state)
- `nA`: number of actions
- `p_diabetes`: probability of a diabetic patient
- `n_bootstrap` : number of bootstrap

````python
config = {'Gamma': 2.0, 'num_itrs': 10000, 'max_horizon': 5, 'discount': 0.99,
          'confounding_threshold': 0.75, 'nS': State.NUM_FULL_STATES + 1, 
          'nA': Action.NUM_ACTIONS_TOTAL, 'p_diabetes': 0.2, 'n_bootstrap': 500}
````

Data generation process uses `t0_policy` for the first timestep, and `mixed_policy` for afterward.

````python
dgen = DGEN.conf_data_generator(transitions=(tx, tr), 
            policies=(mixed_policy, t0_policy), 
            value_fn=value_function, config=config)
trajectories, returns = dgen.simulate(config['num_itrs'], use_tqdm=True)
````

### Weighted Importance Sampling

Weighted Importance Sampling estimate: This section estimates the WIS estimate of each policies `optimal_policy_st`, `with antibiotics` and `without antibiotics`. Each policy consist of two different policies:
1. For time step 0 :
    - `optimal_policy_st` : uses the soft optimal policy
    - `with antibiotics` : administer antibiotics
    - `without antibiotics`: does not administer antibiotics 
2. Afterward: all policies uses `optimal_policy_st`

````python
# make the WIS class
wis = CWIS.conf_wis(trajectories=trajectories, returns=returns, k=1, config=config)
# evaluation policies
with_antibiotics = (t0_policy[0, :, :], optimal_policy_st)
without_antibiotics = (t0_policy[1, :, :], optimal_policy_st)
optimal_policy = (optimal_policy_st, optimal_policy_st)
````

````python
wis_with_anti = wis.compute(with_antibiotics)
wis_without_anti = wis.compute(without_antibiotics)
wis_optimal = wis.compute(optimal_policy)

print("Weighted Importance Sampling Estimates:")
print("With Antibiotics: {}".format(wis_with_anti.mean()))
print("Without Antibiotics: {}".format(wis_without_anti.mean()))
print("Soft Optimal: {}".format(wis_optimal.mean()))
````

### Our method

`core.loss_minimization` can be used to compute our and naive method lower/ upper bound. Here we demonstrate how to compute the lower (and upper) bound. Our methods requires the following inputs:

1. `config` the config file should containts the following information:
    - `Gamma` : amount of confounding
    - `lr` : learning rate
    - `epoch`: number of training epochs
    - `nS`: number of states
    - `nA`: number of actions
    - `bootstrap`: if perform bootstraping
    - `n_bootstrap`: number of bootstraps (only if `bootstrap = True`)
    - `batch_size`: training batchsize, if not specified `batchsize = num_data`
    and example of config is:
    
```python
config = {'Gamma': 2.0, 'lr':5 * 1e-2, 'epoch':80, 'nS': State.NUM_FULL_STATES + 1, 
          'nA': Action.NUM_ACTIONS_TOTAL, 'bootstrap': True, 'n_bootstrap': 500}
```

2. `data` a dictionary of 
    - `samps` : similar to WIS, the output of ConfDataGenerator.simulate(), that contains states, action : `num_data x H x 5`
    - `returns` : similar to WIS, the output of ConfDataGenerator.simulate(), that contains discounted returns : `num_data x 1`
    
3. `eval_policies` : a tuple of evaluation policy
    - `pi_e_t0`: first step policy
    - `pi_e_t1`: policy after the second step
4. `scope`: tensorflow variable scope

````python
config = {'Gamma': 2.0, 'lr':5 * 1e-2, 'epoch':80, 'nS': State.NUM_FULL_STATES + 1, 
          'nA': Action.NUM_ACTIONS_TOTAL, 'bootstrap': True, 'n_bootstrap': 25}
lb_data = {'samps': trajectories, 'returns': returns}
````

````python
with_anti_lb = LB.loss_minimization(config=config, data=lb_data, 
                                   evaluation_policies=with_antibiotics, scope='with_anti_biotics')
without_anti_lb = LB.loss_minimization(config=config, data=lb_data, 
                                   evaluation_policies=without_antibiotics, scope='without_anti_biotics')
optimal_lb = LB.loss_minimization(config=config, data=lb_data, 
                                   evaluation_policies=optimal_policy, scope='optimal')
````

````python
with_anti_lb_ours, with_anti_lb_naive, _ = with_anti_lb.run(use_tqdm=True)
without_anti_lb_ours, without_anti_lb_naive, _ = without_anti_lb.run(use_tqdm=True)
optimal_lb_ours, optimal_lb_naive, _ = optimal_lb.run(use_tqdm=True)
````

````python
print("### Our Method lower bound: Gamma = {}".format(config['Gamma']))
print("With Antibiotics: {:.2f}".format(with_anti_lb_ours.mean()))
print("Without Antibiotics: {:.2f}".format(without_anti_lb_ours.mean()))
print("Soft Optimal: {:.2f}".format(optimal_lb_ours.mean()))

print("\n### Naive lower bound: Gamma = {}".format(config['Gamma']))
print("With Antibiotics: {:.2f}".format(with_anti_lb_naive.mean()))
print("Without Antibiotics: {:.2f}".format(without_anti_lb_naive.mean()))
print("Soft Optimal: {:.2f}".format(optimal_lb_naive.mean()))
````

## Design Sensitivity

In this section, we evaluates the upper and lower bound on two different evaluation policies, `with antibiotics` and `without antibiotics`. The goal is to identify the value of confounding $\Gamma^\prime$ such that if $\Gamma \geq \Gamma^\prime$ we cannot guarantee the superior performance of `with antibiotics`.

Here we generate data with no confounding ($\Gamma^\star = 1.0$), and evaluates both naive and our method. In order to evaluate the upper bound just set the flag `upper_bound` to `True` while calling function `run()`. With this flag, `core.loss_minimization` will evaluate the negative of lowerbound on the negative of the return. 

use function `plot_design_sensitivity` in `utils.utils` to generate similar plots as in the paper. You can find the results of runs in `data/results.zip` as
1. `sens_gamma_5.pkl` : results for $\Gamma^\star=5.0$
2. `sens_gamma_1.pkl` : results for $\Gamma^\star=1.0$

````python
Gamma_Star = 1.0
config = {'Gamma': Gamma_Star, 'num_itrs': 10000, 'max_horizon': 5, 'discount': 0.99,
          'confounding_threshold': 0.75, 'nS': State.NUM_FULL_STATES + 1, 
          'nA': Action.NUM_ACTIONS_TOTAL, 'p_diabetes': 0.2, 'n_bootstrap': 500}
dgen = DGEN.conf_data_generator(transitions=(tx, tr), 
            policies=(mixed_policy, t0_policy), 
            value_fn=value_function, config=config)
trajectories, returns = dgen.simulate(config['num_itrs'], use_tqdm=True)
````

````python
loss_minimization_config = {'Gamma': 1.0, 'lr':5 * 1e-2, 'epoch':150, 
            'nS': State.NUM_FULL_STATES + 1, 'nA': Action.NUM_ACTIONS_TOTAL, 
            'bootstrap': True, 'n_bootstrap': 50}
lb_data = {'samps': trajectories, 'returns': returns}
````

````python
GAMMAs = np.arange(1.0, 7.0, 0.5)
with_results = np.zeros((4, len(GAMMAs), loss_minimization_config['n_bootstrap']))
without_results = np.zeros((4, len(GAMMAs), loss_minimization_config['n_bootstrap']))

with_anti_lb = LB.loss_minimization(config=loss_minimization_config, data=lb_data, 
                evaluation_policies=with_antibiotics, scope='with_anti_biotics')
without_anti_lb = LB.loss_minimization(config=loss_minimization_config, data=lb_data, 
                evaluation_policies=without_antibiotics, scope='without_anti_biotics')
i = 0
for Gamma in tqdm(GAMMAs):
    with_anti_lb.update_gamma(Gamma)
    without_anti_lb.update_gamma(Gamma)
    # Lowerbounds
    with_results[0, i, :], with_results[1, i, :], _ = with_anti_lb.run()
    without_results[0, i, :], without_results[1, i, :], _ = without_anti_lb.run()

    # Upperbounds
    with_results[2, i, :], with_results[3, i, :], _ = with_anti_lb.run(upper_bound=True)
    without_results[2, i, :], without_results[3, i, :], _ = without_anti_lb.run(upper_bound=True)
    i += 1
````

````python
naive_results = {'GAMMAs': GAMMAs, 'pol1_lower':with_results[1, :, :].mean(axis=-1),
                'pol2_lower':without_results[1, :, :].mean(axis=-1),
                'pol1_upper':with_results[3, :, :].mean(axis=-1),
                'pol2_upper':without_results[3, :, :].mean(axis=-1),
                'pol1_label': 'with antibiotics', 
                'pol2_label': 'without antibiotics',
                'cross': 1.23}
````

````python
plt.figure(figsize=(7,5))
plot_design_sensitivity(naive_results, title='Design Sensitivity, Naive', 
                yticks=[-0.1, 0, 0.1, 0.2, 0.3, 0.4],  
                xlim=[0.95, 2.5], ylim=[-0.01, 0.18])
plt.xticks([1.0, 1.5, 2.0, 2.5, 3.0])

#plt.tight_layout()
#plt.savefig("{}/sepsis-sensitivity-naive-{}.pdf".format(figdir, 1))
````

````python
naive_results = {'GAMMAs': GAMMAs, 'pol1_lower':with_results[0, :, :].mean(axis=-1),
                'pol2_lower':without_results[0, :, :].mean(axis=-1),
                'pol1_upper':with_results[2, :, :].mean(axis=-1),
                'pol2_upper':without_results[2, :, :].mean(axis=-1),
                'pol1_label': 'with antibiotics', 
                'pol2_label': 'without antibiotics',
                'cross': 1.73}
````

````python
plt.figure(figsize=(7,5))
plot_design_sensitivity(naive_results, title='Design Sensitivity, Ours', 
                yticks=[-0.1, 0, 0.1, 0.2, 0.3, 0.4],  
                xlim=[0.95, 2.5], ylim=[-0.01, 0.18])
plt.xticks([1.0, 1.5, 2.0, 2.5, 3.0])

#plt.tight_layout()
#plt.savefig("{}/sepsis-sensitivity-naive-{}.pdf".format(figdir, 1))
````
